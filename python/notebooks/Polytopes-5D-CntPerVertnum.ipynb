```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data (vertex_count, unique_polytopes)
data = [
    (6, 49999), (7, 378690), (8, 1856518), (9, 6489509),
    (10, 17779934), (11, 39356823), (12, 72966699), (13, 116420864),
    (14, 163676172), (15, 206729010), (16, 237903244), (17, 252455249),
    (18, 248977476), (19, 229778914), (20, 199440531), (21, 163771967),
    (22, 127881404), (23, 95432370), (24, 68346351), (25, 47123952),
    (26, 31297761), (27, 20024636), (28, 12348822), (29, 7341478),
    (30, 4207677), (31, 2330342), (32, 1248588), (33, 647563),
    (34, 325145), (35, 158938), (36, 75838), (37, 35962),
    (38, 16953), (39, 7641), (40, 3253), (41, 1264),
    (42, 417), (43, 129), (44, 44), (45, 6), (46, 3)
]

vertex_counts_5d_df = pd.DataFrame(data, columns=["vertex_count", "count"])

# Normalize
vertex_counts_5d_df["relative_frequency"] = (
    vertex_counts_5d_df["count"] /
    vertex_counts_5d_df["count"].sum()
)

print(vertex_counts_5d_df[["vertex_count", "count"]])
```

```python
# Plot
plt.figure(figsize=(10, 4))

plt.plot(
    vertex_counts_5d_df["vertex_count"],
    vertex_counts_5d_df["relative_frequency"],
    marker="o",
    linestyle="-"
)

plt.xlabel("Number of vertices")
plt.ylabel("Relative frequency")

plt.tight_layout()
plt.savefig("poly5d-normalized-vertex-dist.png", dpi=300, bbox_inches="tight")
plt.show()
```
