```python
import pandas as pd

# Load the Parquet file
parquet_file = '/run/media/eva/01DA95886B3817A0/Mathfiles/reflexive/ws-5d-reflexive-0000.parquet'
df = pd.read_parquet(parquet_file)

# Ensure the dataframe has at least 6 columns
if df.shape[1] < 6:
    raise ValueError("The dataframe must have at least 6 columns")

# Select the first 6 columns
columns = df.iloc[:, :6]

# Calculate the sum of the first 6 columns
columns['d'] = columns.sum(axis=1)

# Create the desired format
output_df = columns[['d'] + columns.columns.tolist()[:-1]]

# Convert to the required text format
output_file = 'output.txt'
output_df.to_csv(output_file, sep=' ', header=False, index=False)
```
