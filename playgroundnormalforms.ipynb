```python
import numpy as np

def generate_full_rank_matrix(m, n, q):
    # Check if m >= n for full rank
    if m < n:
        raise ValueError("Number of rows (m) must be greater than or equal to number of columns (n) for full rank.")

    # Generate a random matrix with integer entries
    matrix = np.random.randint(-q, q, size=(m, n))

    # Check if rank equals min(m, n), if not, add rows until rank is full
    while np.linalg.matrix_rank(matrix) < min(m, n):
        extra_rows = np.random.randint(-q, q, size=(1, n))
        matrix = np.vstack([matrix, extra_rows])
    return matrix

# Example usage:
m = 6
n = 5
q=30

with open("/content/randompoly_nf500.txt", "w") as f:

  for i in range(100):

    random_full_rank_matrix = generate_full_rank_matrix(m, n,q)
    p = Polytope(random_full_rank_matrix)
    nf = p.normal_form(affine_transform=True,backend="native")
    f.write(str(nf))
    f.write('\n')
```

```python
import re

# Function to extract matrices from text file
def extract_matrices(file_path):
    with open(file_path, 'r') as file:
        data = file.read()
        # Use regular expression to find matrices
        matrices = re.findall(r'\[\[(.*?)\]\]', data, re.DOTALL)
        return matrices

# Function to calculate statistics for the last row of each matrix
def calculate_statistics(matrices):
    statistics = []
    for matrix in matrices:
        rows = matrix.strip().split('\n')
        last_row = [int(num) for num in rows[-1].strip().split()]
        min_val = min(last_row)
        max_val = max(last_row)
        statistics.append({'min': min_val, 'max': max_val})
    return statistics

# Main function
def main():
    file_path = '/content/randompoly_nf.txt'  # Replace with your file path
    matrices = extract_matrices(file_path)
    print(matrices)
    if matrices:
        statistics = calculate_statistics(matrices)
        for i, stats in enumerate(statistics, start=1):
            print(f"Matrix {i}:")
            print(f"  Min value of last row: {stats['min']}")
            print(f"  Max value of last row: {stats['max']}")
    else:
        print("No matrices found in the file.")

if __name__ == "__main__":
    main()
```
